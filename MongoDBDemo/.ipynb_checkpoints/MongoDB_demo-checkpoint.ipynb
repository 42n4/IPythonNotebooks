```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pymongo
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
db = client.test
collection = db.people
```

```python
collection.drop()
os.system('mongoimport -d test -c people dummyData.json')
```

```python
cursor = collection.find().sort('Age',pymongo.ASCENDING).limit(3)
for doc in cursor:
    print doc
```

```python
pipeline = [
        {"$group": {"_id":"$Country",
             "AvgAge":{"$avg":"$Age"},
             "Count":{"$sum":1},
        }},
        {"$sort":{"Count":-1,"AvgAge":1}}
]
aggResult = collection.aggregate(pipeline)["result"]
df1 = pd.DataFrame(aggResult)
df1 = df1.set_index("_id")
df1[:5]
```

```python
pipeline = [
        {"$match": {"Country":"China"}},
]
aggResult = collection.aggregate(pipeline)["result"]
df2 = pd.DataFrame(aggResult)
df2[:5]
```

```python
from IPython.display import HTML
import folium
import numpy as np

def inline_map(map):
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))
```

```python
world_map = folium.Map(location=[35, 100], 
                    zoom_start=4)
for i in range(len(df2)):
    world_map.simple_marker(location=df2.Location[i].split(','), popup=df2.Name[i]+', age:'+str(df2.Age[i]))
inline_map(world_map)
```
